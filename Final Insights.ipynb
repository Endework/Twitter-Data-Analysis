```python
import pandas as pd
data=pd.read_csv(r'C:\Users\ende\Desktop\Twitter-Data-Analysis-main (1)\processed_tweet_data.csv')
data.head()
```

```python
import string
def punctuation_removal(text):
    all_list = [char for char in text if char not in string.punctuation]
    clean_str = ''.join(all_list)
    return clean_str
data['cleaned_text'] = data['text'].apply(punctuation_removal)
```

```python
import nltk
from nltk.corpus import stopwords
stop = stopwords.words('english')
data['cleaned_text'] = data['cleaned_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
data = data.query("lang == 'en' ")
data.head(10)
```

```python
# take the rows from the hashtag columns where there are actually hashtags
hashtags_list_df = data.loc[
                       data.hashtags.apply(
                           lambda hashtags_list: hashtags_list !=[]
                       ),['hashtags']]
```

```python
hashtags_list_df
```

```python
 #create dataframe where each use of hashtag gets its own row
flattened_hashtags_df = pd.DataFrame(
    [hashtag for hashtags_list in hashtags_list_df.hashtags
    for hashtag in hashtags_list],
    columns=['hashtag'])
```

```python
flattened_hashtags_df
```

```python
#add flatten_hashtags to tweet_df
data["flattened_hashtags"]= flattened_hashtags_df
```

```python
# To plot Top 20 Hashtags
data['flattened_hashtags'].value_counts()[:15].plot(kind='barh',color='orange',title='flattend hashtags',xlabel='hashtags',ylabel='value count')
```
