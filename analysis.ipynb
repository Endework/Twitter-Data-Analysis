```python
import pandas as pd
data=pd.read_csv(r'C:\Users\ende\Desktop\Twitter-Data-Analysis-main (1)\processed_tweet_data.csv')
data.head()
```

```python
import string
def punctuation_removal(text):
    all_list = [char for char in text if char not in string.punctuation]
    clean_str = ''.join(all_list)
    return clean_str
data['cleaned_text'] = data['text'].apply(punctuation_removal)
```

```python
import nltk
from nltk.corpus import stopwords
stop = stopwords.words('english')
data['cleaned_text'] = data['cleaned_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
import numpy as np
conditions = [
    (data['sentiment'] < 0),
    (data['sentiment'] >= 0)
     ]
  
    

values = ['negative', 'positive']

data['label'] = np.select(conditions, values)

data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data['cleaned_text'], data.label, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn import metrics
import itertools
```

```python

pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', LogisticRegression())])
model = pipe.fit(X_train, y_train)
prediction = model.predict(X_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction)*100,2)))
cm = metrics.confusion_matrix(y_test, prediction)
plot_confusion_matrix(cm, classes=['positive', 'negative'])
```
